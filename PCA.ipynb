```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_diabetes

# Load the diabetes dataset
diabetes = load_diabetes()

# Convert the dataset to a pandas DataFrame
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Apply PCA to reduce dimensionality to 2 components
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Check the shape of the reduced data
print("Original shape:", data.shape)
print("Reduced shape after PCA:", data_pca.shape)

# Explained variance ratio to see how much variance is retained
print("Explained variance ratio of components:", pca.explained_variance_ratio_)
print("Total variance retained:", sum(pca.explained_variance_ratio_))

# Plot the PCA result
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA on Diabetes Dataset (2 Components)')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine

# Step 1: Load the Wine dataset (or use any other large dataset)
wine = load_wine()

# Convert it to a DataFrame for better manipulation
data = pd.DataFrame(wine.data, columns=wine.feature_names)
target = wine.target

# Step 2: Standardize the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 3: Apply PCA for dimensionality reduction
# Let's reduce the dimensionality to 2 components
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)

# Check explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by each component: {explained_variance}')
print(f'Total variance explained by the first 2 components: {sum(explained_variance)}')

# Step 4: Visualize the PCA result
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=target, cmap='plasma', edgecolor='k', s=100)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA on Wine Dataset (2 Components)')
plt.colorbar(label='Target Label')
plt.show()

# Step 5: Plot Cumulative Explained Variance to decide on optimal number of components
pca_full = PCA().fit(data_scaled)
cumulative_variance = np.cumsum(pca_full.explained_variance_ratio_)

plt.figure(figsize=(8, 6))
plt.plot(np.arange(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Step 1: Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target  # Add target column to the DataFrame

# Step 2: Standardize the features (PCA works better with standardized data)
features = iris.feature_names  # List of features (columns)
x = df.loc[:, features].values  # Extracting the features
y = df.loc[:, ['target']].values  # Extracting the target (class labels)

# Standardize the data (mean = 0, variance = 1)
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Step 3: Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reduce to 2 principal components
principal_components = pca.fit_transform(x_scaled)

# Step 4: Create a DataFrame with the principal components
df_pca = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['target'] = df['target']

# Step 5: Visualize the result
plt.figure(figsize=(8, 6))
targets = [0,1,2]  # The three species in the Iris dataset
colors = ['r', 'g', 'b']
labels = iris.target_names  # Setosa, Versicolor, Virginica

for target, color, label in zip(targets, colors, labels):
    indices = df_pca['target'] == target
    plt.scatter(df_pca.loc[indices, 'Principal Component 1'],
                df_pca.loc[indices, 'Principal Component 2'],
                c=color, s=100, edgecolor='k', label=label)

plt.title('PCA on Iris Dataset (2 Principal Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(labels)
plt.grid(True)
plt.show()

# Step 6: Check explained variance by each component
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by each component: {explained_variance}')
print(f'Total variance explained by the first 2 components: {sum(explained_variance)}')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris, load_wine, load_breast_cancer
from sklearn.preprocessing import StandardScaler

# Function to apply PCA and visualize the result
def apply_pca(dataset, dataset_name, n_components=2):
    # Step 1: Load the dataset and prepare DataFrame
    df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
    df['target'] = dataset.target

    # Step 2: Standardize the features
    x = df.drop('target', axis=1)  # Remove target column to keep only features
    scaler = StandardScaler()
    x_scaled = scaler.fit_transform(x)

    # Step 3: Apply PCA
    pca = PCA(n_components=n_components)
    principal_components = pca.fit_transform(x_scaled)

    # Step 4: Create DataFrame for PCA components
    df_pca = pd.DataFrame(data=principal_components,
                          columns=[f'PC{i+1}' for i in range(n_components)])
    df_pca['target'] = df['target']

    # Step 5: Visualize PCA result (first two components)
    plt.figure(figsize=(8, 6))
    targets = np.unique(df['target'])
    colors = ['r', 'g', 'b', 'c', 'm', 'y']

    for target, color in zip(targets, colors[:len(targets)]):
        indices = df_pca['target'] == target
        plt.scatter(df_pca.loc[indices, 'PC1'],
                    df_pca.loc[indices, 'PC2'],
                    c=color, s=100, edgecolor='k', label=f'Class {target}')

    plt.title(f'PCA on {dataset_name} (First 2 Components)')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.legend(loc='best')
    plt.grid(True)
    plt.show()

    # Step 6: Explained variance
    explained_variance = pca.explained_variance_ratio_
    cumulative_variance = np.cumsum(explained_variance)
    print(f'Explained variance by each component (PCA on {dataset_name}): {explained_variance}')
    print(f'Total variance explained by first {n_components} components: {cumulative_variance[-1]}')

    # Plotting cumulative explained variance
    plt.figure(figsize=(8, 6))
    plt.plot(np.arange(1, len(explained_variance) + 1), cumulative_variance, marker='o', linestyle='--')
    plt.title(f'Cumulative Explained Variance on {dataset_name}')
    plt.xlabel('Number of Principal Components')
    plt.ylabel('Cumulative Explained Variance')
    plt.grid(True)
    plt.show()


# 1. Iris Dataset
iris = load_iris()
apply_pca(iris, 'Iris Dataset')

# 2. Wine Dataset
wine = load_wine()
apply_pca(wine, 'Wine Dataset')

# 3. Breast Cancer Dataset
breast_cancer = load_breast_cancer()
apply_pca(breast_cancer, 'Breast Cancer Dataset')
```
