```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift
from sklearn.mixture import GaussianMixture
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the Diabetes dataset
diabetes = load_diabetes()
X = diabetes.data

# Standardize the data (important for many clustering algorithms)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Optional: Use PCA to reduce dimensionality for easier visualization (2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Plotting function to visualize clustering results
def plot_clusters(X, labels, title):
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=50)
    plt.title(title)
    plt.show()

# 1. K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
plot_clusters(X_pca, kmeans_labels, "K-Means Clustering")

# 2. Agglomerative Hierarchical Clustering
agglo = AgglomerativeClustering(n_clusters=3)
agglo_labels = agglo.fit_predict(X_scaled)
plot_clusters(X_pca, agglo_labels, "Agglomerative Clustering")

# Plot Dendrogram for Hierarchical Clustering
Z = linkage(X_scaled, 'ward')
plt.figure()
dendrogram(Z)
plt.title("Dendrogram for Agglomerative Clustering")
plt.show()

# 3. DBSCAN Clustering
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
plot_clusters(X_pca, dbscan_labels, "DBSCAN Clustering")

# 4. Mean Shift Clustering
meanshift = MeanShift()
meanshift_labels = meanshift.fit_predict(X_scaled)
plot_clusters(X_pca, meanshift_labels, "Mean Shift Clustering")

# 5. Gaussian Mixture Model (GMM) Clustering
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X_scaled)
plot_clusters(X_pca, gmm_labels, "Gaussian Mixture Model (GMM) Clustering")
```

# Hierarchical clustering

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features
y = iris.target  # True labels (for reference)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Hierarchical Clustering
linked = linkage(X_scaled, method='ward')

# Create a dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linked, orientation='top', labels=iris.target_names[y], distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Iris Species')
plt.ylabel('Euclidean Distance')
plt.show()

# Cluster the data (choose a threshold for clusters)
threshold = 5 # Adjust this value based on dendrogram
clusters = fcluster(linked, threshold, criterion='distance')

# Visualize the clustering results in 2D (using the first two features)
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title('Hierarchical Clustering of Iris Dataset (2D Visualization)')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.colorbar(label='Cluster Label')
plt.grid()
plt.show()
```

```python
# DBSCAN code
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)  # Adjust eps and min_samples as needed
labels = dbscan.fit_predict(X_scaled)

# Print the number of clusters and noise points
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Exclude noise
n_noise = list(labels).count(-1)  # Count noise points
print(f'Estimated number of clusters: {n_clusters}')
print(f'Estimated number of noise points: {n_noise}')

# Visualize the results
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = plt.cm.get_cmap('viridis', len(unique_labels))

for k, col in zip(unique_labels, colors.colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)
    plt.scatter(X_scaled[class_member_mask, 0], X_scaled[class_member_mask, 1],
                color=col, edgecolor='k', s=50, label=f'Cluster {k}')

plt.title('DBSCAN Clustering on Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define grid parameters
grid_size = 0.5
x_min, x_max = X_scaled[:, 0].min(), X_scaled[:, 0].max()
y_min, y_max = X_scaled[:, 1].min(), X_scaled[:, 1].max()

# Create grid
x_bins = np.arange(x_min, x_max + grid_size, grid_size)
y_bins = np.arange(y_min, y_max + grid_size, grid_size)
grid = np.zeros((len(x_bins) - 1, len(y_bins) - 1))

# Assign points to grid cells
for point in X_scaled:
    x_idx = np.digitize(point[0], x_bins) - 1
    y_idx = np.digitize(point[1], y_bins) - 1
    if 0 <= x_idx < grid.shape[0] and 0 <= y_idx < grid.shape[1]:
        grid[x_idx, y_idx] += 1  # Increment count for the grid cell

# Find dense grid cells
threshold = 2  # Minimum number of points required in a grid cell to be considered dense
dense_cells = np.where(grid >= threshold)

# Create a DataFrame to hold points that fall into dense cells
dense_points = []
for i in range(len(dense_cells[0])):
    x_bin = dense_cells[0][i]
    y_bin = dense_cells[1][i]
    x_range = (x_bins[x_bin], x_bins[x_bin + 1])
    y_range = (y_bins[y_bin], y_bins[y_bin + 1])
    for point in X_scaled:
        if x_range[0] <= point[0] < x_range[1] and y_range[0] <= point[1] < y_range[1]:
            dense_points.append(point)

# Convert dense points to a numpy array for visualization
dense_points = np.array(dense_points)

# Plotting results
plt.figure(figsize=(10, 6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], color='lightgray', label='Data Points')
plt.scatter(dense_points[:, 0], dense_points[:, 1], color='orange', label='Dense Points')
plt.title('Grid-Based Clustering on Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Fit a Gaussian Mixture Model
n_components = 3  # Number of clusters
gmm = GaussianMixture(n_components=n_components, random_state=42)
gmm.fit(X_scaled)

# Predict the cluster labels
labels = gmm.predict(X_scaled)

# Print the parameters of the GMM
print("Weights of each Gaussian component:\n", gmm.weights_)
print("Means of each Gaussian component:\n", gmm.means_)
print("Covariances of each Gaussian component:\n", gmm.covariances_)

# Visualize the results
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = plt.cm.get_cmap('viridis', len(unique_labels))

for k, col in zip(unique_labels, colors.colors):
    class_member_mask = (labels == k)
    plt.scatter(X_scaled[class_member_mask, 0], X_scaled[class_member_mask, 1],
                color=col, edgecolor='k', s=50, label=f'Cluster {k}')

plt.title('Gaussian Mixture Model Clustering on Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import SpectralClustering

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Spectral Clustering
n_clusters = 3  # Number of clusters
spectral = SpectralClustering(n_clusters=n_clusters, affinity='nearest_neighbors', random_state=42)
labels = spectral.fit_predict(X_scaled)

# Visualize the results
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = plt.cm.get_cmap('viridis', len(unique_labels))

for k, col in zip(unique_labels, colors.colors):
    class_member_mask = (labels == k)
    plt.scatter(X_scaled[class_member_mask, 0], X_scaled[class_member_mask, 1],
                color=col, edgecolor='k', s=50, label=f'Cluster {k}')

plt.title('Spectral Clustering on Iris Dataset')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from tslearn.clustering import TimeSeriesKMeans
from tslearn.preprocessing import TimeSeriesScalerMinMax

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target

# Create synthetic time series data based on the Iris dataset
# For example, we can create time series for each species by repeating the measurements over time
# Let's assume each species has 5 time points (for example, 5 measurements taken over time)
time_points = 5
n_species = len(np.unique(target))
time_series_data = []

for i in range(n_species):
    # Repeat the measurements for the species
    species_data = np.tile(data[target == i], (time_points, 1)).T
    time_series_data.append(species_data)

# Convert the list to a 3D numpy array (samples, time points, features)
time_series_data = np.array(time_series_data)

# Scale the time series data
scaler = TimeSeriesScalerMinMax()
time_series_scaled = scaler.fit_transform(time_series_data)

# Apply K-Shape clustering
n_clusters = n_species  # Set number of clusters to the number of species
model = TimeSeriesKMeans(n_clusters=n_clusters, metric="euclidean", random_state=0)
labels = model.fit_predict(time_series_scaled)

# Visualize the results
plt.figure(figsize=(12, 8))
for yi in range(n_clusters):
    plt.subplot(n_clusters, 1, yi + 1)
    for xx in time_series_scaled[labels == yi]:
        plt.plot(xx.ravel(), "k-", alpha=0.5)
    plt.plot(model.cluster_centers_[yi].ravel(), "r-", label="Centroid")
    plt.title(f"Cluster {yi + 1} (Species {yi})")
    plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Create a synthetic dataset
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)

# Visualize the synthetic dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Synthetic Dataset for Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Define must-link and cannot-link constraints
# Here we define two must-link constraints and two cannot-link constraints
must_link = [(0, 1), (3, 4)]  # Must-link pairs
cannot_link = [(2, 5), (8, 9)]  # Cannot-link pairs

# K-means clustering with constraints
class ConstraintKMeans:
    def __init__(self, n_clusters, max_iter=100):
        self.n_clusters = n_clusters
        self.max_iter = max_iter

    def fit(self, X):
        self.kmeans = KMeans(n_clusters=self.n_clusters, max_iter=self.max_iter, random_state=0)
        self.labels_ = self.kmeans.fit_predict(X)

        # Apply must-link constraints
        for (i, j) in must_link:
            if self.labels_[i] != self.labels_[j]:
                # Force them to be in the same cluster
                self.labels_[j] = self.labels_[i]

        # Apply cannot-link constraints
        for (i, j) in cannot_link:
            if self.labels_[i] == self.labels_[j]:
                # Force them to be in different clusters
                new_label = max(self.labels_) + 1
                self.labels_[j] = new_label

        return self.labels_

# Apply the constraint-based K-means
constraint_kmeans = ConstraintKMeans(n_clusters=3)
labels = constraint_kmeans.fit(X)

# Visualize the results
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.title('Constraint-Based Clustering Results')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
